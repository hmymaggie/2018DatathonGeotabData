```python
%matplotlib inline
import pandas as pd
from IPython.display import display, Markdown
pd.options.display.max_columns = 100
```

```python
def describe_data(df):
    display(Markdown("**Data Shape**: %s" % str(df.shape)))
    display(Markdown("<br>**Column names**:"))
    display(df.columns.values)
    display(Markdown("<br>**Sample**:"))
    display(df.head())
    display(Markdown("<br>**Statistics**:"))
    display(df.describe())
```

# Toronto Police Killed and Serious Accidents Dataset

```python
describe_data(pd.read_csv('ksi.csv'))
```

# Geotab Hazardous Driving Areas Dataset

```python
describe_data(pd.read_csv('hazardous_driving.csv'))
```

# Geotab Road Impediments Dataset

```python
describe_data(pd.read_csv('roadimpediments.csv'))
```

# City of Toronto One Address Repository

```python
describe_data(pd.read_csv('toronto_addresses.csv'))
```

# City of Toronto Centerline Intersection

```python
describe_data(pd.read_csv('toronto_roads.csv'))
```
